## Cleaning

```python
backup=car.copy()
```

```python
car=car[car['year'].str.isnumeric()]
```

```python
car['year'].str.isnumeric()
```

```python
car['year']=car['year'].astype(int)
```

```python
car.info()
```

```python
car['Price']!="Ask For Price"
```

```python
car=car[car['Price']!="Ask For Price"]
```

```python
car['Price']=car['Price'].str.replace(',','').astype(int)
```

```python
car.info()
```

```python
car['kms_driven']=car['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
car=car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven']=car['kms_driven'].astype(int)
```

```python
car.info()
```

```python
car=car[~car['fuel_type'].isna()]
```

```python
car[car['fuel_type'].isna()]
```

```python
# Ensure all values in 'name' are strings
car['name'] = car['name'].apply(lambda val: ' '.join(val) if isinstance(val, list) else str(val))

# Split the string, slice the first three parts, and join them back
car['name'] = car['name'].str.split().str.slice(start=0, stop=3).str.join(' ')

print(car)
```

```python
x=car.drop(columns='Price')
y=car['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
car['name'].unique()
```

```python
ohe.fit(x[['name', 'company', 'fuel_type']])
```

```python
ohe.categories_
```

```python
#Creating a column transformer to transform categorical columns

column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python

pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
import numpy as np
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe=make_pipeline(column_trans,lr)
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
r2_score(y_test,y_pred)
```
